```python
def get_location_account_mapping(df):
    """
    Extracts unique mappings of 'Location_Name' to 'Account ID'.
    
    @param df: The input DataFrame containing 'Location_Name' and 'Account ID'.
    @return: Dictionary mapping 'Location_Name' to 'Account ID'.
    """
    # Drop rows where Account ID or Location Name is missing
    mapping_df = df[["Location_Name", "Account_ID"]].dropna().drop_duplicates()
    
    # Convert to dictionary
    location_account_mapping = dict(zip(mapping_df["Location_Name"], mapping_df["Account_ID"]))
    
    return location_account_mapping

# Example usage
location_account_mapping = get_location_account_mapping(df_beacon)
print(location_account_mapping)
```

<h1> Parsing City of Houston (CoH) data </h1>

```python
WEST_CAMPUS_GROUP = [
    "Anderson Clarke",
    "Football Stadium"
]

LIBRARY_SERVICE_CENTER = [
    "Library Service Ctr",
]

NORTH_PLANT_GROUP = [
    "Main Feed #1 Current chg",
    "Water Well",
    "Martel College"
]

coh_groups = {
    "west_campus_group" : WEST_CAMPUS_GROUP,
    "library_service_center": LIBRARY_SERVICE_CENTER,
    "north_plant_group": NORTH_PLANT_GROUP,
    
}
```

```python
def parse_coh_groups(groups):
    # Initialize an empty DataFrame
    final_df = pd.DataFrame()

    for group in groups:

        group_df = pd.DataFrame()
        for name in groups[group]:
            filtered_df = df_coh[df_coh['Old Name'] == name]  # Filter rows matching 'Old Name'
            group_df = pd.concat([group_df, filtered_df], ignore_index=True)

        grouped_df = group_df.groupby('Usage Date').sum()
        grouped_df['Old Name'] = group

        grouped_df.index = pd.to_datetime(grouped_df.index)

        final_df = pd.concat([final_df, grouped_df], ignore_index=False)

    sorted_df = final_df.sort_index()
    return sorted_df
```

```python
parsed_df = parse_coh_groups(coh_groups)
```

<h1> Parsing Beacon Data </h1>

```python
BEACON_LIBRARY_GROUP = {

}

BEACON_WEST_CAMPUS_GROUP = {
    "Anderson-Clarke Center": [98845],
    "Brian Patterson Sports Performance Center": [98818],
    "Brockman Hall for Opera":[98799],
    "Moody Center For The Arts":[98848],
    "Reckling Park At Cameron Field":[98804, 98800],
    "Rice Stadium":[98809, 98811, 98814, 98810, 98812],
    "Tennis Center":[98803],
    "Track & Soccer Stadium":[98816],
    "Tudor Fieldhouse And Youngkin Center":[98865],
}

BEACON_NORTH_PLANT_GROUP = {
    
}

beacon_groups = {
    "west_campus_group" : BEACON_WEST_CAMPUS_GROUP,
}
```

```python

def parse_beacon_groups(groups, df):
    final_df = pd.DataFrame()

    # Loop through each beacon group
    for group_name, group_data in beacon_groups.items():
        group_df = pd.DataFrame()

        # Loop through each building and associated Account IDs
        for location_name, account_ids in group_data.items():
            # Filter rows for the Account IDs associated with the current location
            filtered_df = df[df["Account_ID"].isin(account_ids)]

            # Group by Read_Time and sum the 'Read' column (flows)
            grouped_df = filtered_df.groupby('Read_Time')['Flow'].sum()

            # Create a new column with the location name for each row
            grouped_df = grouped_df.reset_index()
            grouped_df['Location_Name'] = location_name

            group_df = pd.concat([group_df, grouped_df], ignore_index=True)

        final_df = pd.concat([final_df, group_df], ignore_index=True)

    final_df = final_df.sort_values(by='Read_Time')
    final_df = final_df.rename(columns={"Flow": "beacon_flow"})

    return final_df

```

```python
beacon_df = parse_beacon_groups(beacon_groups, df_beacon)
```

<h1> Portioning Flows </h1>

```python

def portion(buildings, coh_agg):
    
    portion_buildings = []

    result_df = pd.DataFrame(index=buildings, columns=coh_agg.index)

    # Iterate by month so we can adjust accordingly
    for index, row in coh_agg.iterrows():
        for b in buildings:
            has_beacon = beacon_df.loc[(beacon_df['Read_Time'] == index) & (beacon_df['Location_Name'] == b)]

            # Has beacon data, add entry, subtract flow from COH data
            if not has_beacon.empty:
                result_df.loc[b, index] = has_beacon['beacon_flow'].values[0]
            
            # No beacon data, must portion
            else:
                portion_buildings.append(b)

        # Compute total GSF
        total_gsf = sum(float(gsf_df[gsf_df['Building Name'] == b]["GSF"]) for b in portion_buildings)
        
        # Compute percentage allocation for each building
        percentages = {b: float(gsf_df[gsf_df['Building Name'] == b]["GSF"]) / total_gsf for b in portion_buildings}
        
        # Distribute 'Gallons' based on percentage
        for b in portion_buildings:
            result_df.loc[b, index] = coh_agg.loc[index, "Gallons"] * percentages[b]

    return result_df
```
